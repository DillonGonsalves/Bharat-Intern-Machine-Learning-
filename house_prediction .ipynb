```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("C:/Users/gonsa/OneDrive/Desktop/Machine Learning/Activity/House Prediction/1553768847_housing.xlsx")
```

```python
data.info()
```

```python
data.dropna(inplace = True) #droping null values
```

```python
#splitting data

from sklearn.model_selection import train_test_split 

X = data.drop(['median_house_value'], axis = 1) #without target variable
y = data['median_house_value'] #target
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,6))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot = True)
```

```python
#Data PreProcessing 


#we did this to remove the skewness observed in above graphs
train_data['total_rooms'] = np.log(train_data['total_rooms']+1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms']+1)
train_data['population'] = np.log(train_data['population']+1)
train_data['households'] = np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
#Encode categorical data : Convert categorical column in the dataset to numerical data.

train_data.ocean_proximity.value_counts()
```

```python
# we will add diff catergory of each data as 1,0 ie yes or no 

train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)
```

```python
train_data
```

```python
#feature eng
```

```python
train_data ['bedroom.ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household_rooms']=train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot = True)
```

```python
from sklearn.linear_model import LinearRegression

X_train,y_train = train_data.drop(['median_house_value'],axis=1), train_data['median_house_value']

reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
#testing
test_data = X_test.join(y_test)
#we did this to remove the skewness observed in above graphs
test_data['total_rooms'] = np.log(test_data['total_rooms']+1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms']+1)
test_data['population'] = np.log(test_data['population']+1)
test_data['households'] = np.log(test_data['households']+1)

# we will add diff catergory of each data as 1,0 ie yes or no 
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

test_data ['bedroom.ratio']=test_data['total_bedrooms']/test_data['total_rooms']
test_data['household_rooms']=test_data['total_rooms']/test_data['households']
```

```python
test_data
```

```python
X_test,y_test = test_data.drop(['median_house_value'],axis=1), test_data['median_house_value']
```

```python
reg.score(X_test, y_test)
```

```python
#random forest Ensemble

from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(X_train, y_train)
```

```python
forest.score(X_test, y_test)
```

```python
#cross validation 

from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [200],
    "max_depth":[None],
    "min_samples_split": [2]
}

grid_search = GridSearchCV(forest, param_grid, cv=5,scoring="neg_mean_squared_error",return_train_score=True)
grid_search.fit(X_train, y_train)
```

```python
best_forest=grid_search.best_estimator_
```

```python
best_forest.score(X_test,y_test)
```
